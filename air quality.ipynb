```python
df=pd.DataFrame(data)
df.drop('Date',axis=1,inplace=True)
```

```python
df.drop('Time',axis=1,inplace=True)
```

```python
X_int=df.drop('CO_level',axis=1).values
Y_int=df['CO_level'].values
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_int,Y_int, test_size=0.3,random_state=42)
```

```python
from sklearn.ensemble import RandomForestClassifier
for i in range(100,200):
    clf=RandomForestClassifier(n_estimators=i,max_depth=2,random_state=13)
    clf.fit(X_train,y_train)
    clf.predict(X_test)
    scr=clf.score(X_test,y_test)
    print(scr)
```

```python
from sklearn.ensemble import RandomForestClassifier

clf=RandomForestClassifier(n_estimators=500,max_depth=6,random_state=13)
clf.fit(X_train,y_train)
clf.predict(X_test)
scr=clf.score(X_test,y_test)
print(scr)
```
